```python
#from qiskit_aer import Aer
 
from qiskit import *
from qiskit.visualization import plot_histogram
import qiskit.qasm2
```

```python
qc=QuantumCircuit(5,1)
qc.h(2)
qc.cx(2,0)
#qc.cx(2,5)
qc.cx(2,1)
qc.barrier()
qc.h([2,3])
qc.cx(3,2)
qc.h([2,3])
qc.barrier()
qc.cx(2,4)
qc.cx(2,0)
#qc.cx(2,5)
qc.barrier()
qc.measure(0,0)
#qc.measure(1,1)
qc.draw(output='mpl')
```

```python
first_circuit = qiskit.qasm2.load("first_circuit.qasm")
```

```python
from qiskit.providers.basic_provider import BasicSimulator
backend = BasicSimulator()
result = backend.run(first_circuit).result()
```

```python
counts  = result.get_counts(first_circuit)
print(counts)
```

```python
plot_histogram(counts)
```
